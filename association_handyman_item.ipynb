```python
import pandas as pd
df = pd.read_csv('Sales_data.csv')
df.head()
```

```python
df = df.drop_duplicates(subset=['Date','StoreNumber', 'CashRegisterNumber', 'ReceiptNumber', 'ArticleName'], keep='first')
display(df.head())
display(df.shape)
```

```python
display(df[df.ArticleName.str.contains("Werkzeug")].head())
df[df.ArticleName.str.contains("Werkzeug")].shape
```

```python
df.loc[:,"ArticleName"][df.ArticleName.str.contains("Werkzeug")] = "Werkzeug"
```

```python
list_ = []
for ind, row in df[df.ArticleName=='Werkzeug'].iterrows():
    list_.append([row.Date, row.StoreNumber, row.CashRegisterNumber, row.ReceiptNumber])
```

```python
list_[0:5]
```

```python
trans = pd.DataFrame()
for list in list_:
    Date = list[0]
    StoreNumber = list[1]
    CashRegisterNumber = list[2]
    ReceiptNumber = list[3]
    result = df.query(f"Date=='{Date}' & StoreNumber=={StoreNumber} & CashRegisterNumber=={CashRegisterNumber} & ReceiptNumber=={ReceiptNumber}")
    trans = pd.concat([trans, result], ignore_index=True)
display(trans.head())
```

```python
def set_group(x):
    if x['ArticleName'] in ['Pfandartikel (Tandem) - EW', 'Pfandartikel (Tandem) - MW']:
        return "Pfandartikel"
    if x['ArticleName'] in ['BO-Schnittbr.1ST', 'BO-Schnittbroetchen 1ST', 'BO-Kaiserbroetchen 1ST', 'BO-Kaiserbroetchen 65g']:
        return "Broetchen"
    if x['ArticleName'] in ['GL H-Milch1,5%1L VLOG', 'GL H-Milch 3,5% 1L VLOG']:
        return "Milch"
    else:
        return x['ArticleName']

trans['Group'] = trans.apply(set_group, axis=1)
```

```python
trans.shape
```

```python
trans.Group.value_counts()[0:10]
```

```python
# drop articles with overwhelming quantity
trans.drop(trans[trans.Group == "Pfandartikel"].index, inplace=True)
trans.head()
```

```python
# integrate the identical transactions
integrated = trans.groupby(['Date','StoreNumber', 'CashRegisterNumber', 'ReceiptNumber']).agg(Articles=('Group', lambda x:';'.join(x.unique()))).reset_index()
display(integrated.head())

# spilt the integrated transactions into several columns
split_ = integrated['Articles'].str.split(';', expand=True).dropna(how='all', axis=1)
display(split_.head())

for i in range(6):
    mini_article = split_[split_.iloc[:,i].notnull()]
    print(f"The minimum article is {i+1} and the frequency is {mini_article.shape[0]}" )

# filter transactions with at least 4 articles
split_ = split_[split_.iloc[:,2].notnull()]
display(split_.head())

# convert DataFrame to List
# delete the NoneType
res = []
def filter_list(_list):
    for item in _list:
        res.append([x for x in item if x is not None])
filter_list(split_.values.tolist())

transactions_from_df = [tuple(row) for row in res]
transactions_from_df[0:3]

# apriori algorithm
from efficient_apriori import apriori
itemsets, rules = apriori(transactions_from_df, min_support=0.05, min_confidence=0.5)
print("itemsets:",itemsets)
print("rules", rules)
```
