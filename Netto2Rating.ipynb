```python
# load libraries
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter(action="ignore", category=Warning)
```

```python
df = pd.read_csv("Article_top_Data.csv")
df.head()
```

```python
df = df.drop_duplicates(subset=['Date','StoreNumber', 'CashRegisterNumber', 'ReceiptNumber', 'Group'], keep='first')
df.head()
```

```python
# drop articles with overwhelming quantity
df.drop(df[df.Group == "Pfandartikel"].index, inplace=True)
df.head()
```

```python
df['Duplicates'] = df.groupby(['Date', 'StoreNumber', 'CashRegisterNumber', 'ReceiptNumber'])['Date'].transform('count')
df = df[df['Duplicates'] >= 4]
df.drop(columns=['Duplicates'], inplace=True)
df.head()
```

```python
# 将Date、StoreNumber、CashRegisterNumber、ReceiptNumber相同的行合并，表示同一个用户的所有购买记录
grouped_df = df.groupby(['Date', 'StoreNumber', 'CashRegisterNumber', 'ReceiptNumber']).agg({'Group': lambda x: list(x)})
grouped_df.reset_index(inplace=True)
grouped_df.head()
```

```python
# 构建推荐系统算法所需的格式
rows = []
for index, row in grouped_df.iterrows():
    user_id = str(row['Date']) + '_' + str(row['StoreNumber']) + '_' + str(row['CashRegisterNumber']) + '_' + str(row['ReceiptNumber'])
    items = row['Group']
    for item in items:
        rows.append([user_id, item, 1])

rows[0:5]
```

```python
df = pd.DataFrame(rows, columns=['User', 'Item', 'Rating'])
df.head()
```

```python
df.shape
```

```python
# 将未购买的商品评分设为0
all_items = set(df['Item'].unique())
for user in df['User'].unique():
    user_items = set(df[df['User'] == user]['Item'])
    for item in all_items - user_items:
        df = df.append({'User': user, 'Item': item, 'Rating': 0}, ignore_index=True)
df.head()
```

```python
df.to_csv("rating_data.csv")
```
